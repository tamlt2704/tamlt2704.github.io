```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
%matplotlib inline
```

```python
sankey = Sankey()
sankey.add(flows=[1,-1], labels=['input', 'output'])
sankey.finish()
```

```python
# rotation
sankey = Sankey()
sankey.add(flows=[1,-1], labels=['input', 'output'], rotation=-90)
sankey.finish()
```

```python
sankey = Sankey()
sankey.add(flows=[1, -1, 0.5, -0.5], 
           labels=['input', 'output', 'another input', '4th'],
           orientations=[0, 0, 1, 1], 
           rotation=-90
        )
sankey.finish()
```

```python
# connect sankey diagrams together
```

```python
sankey = Sankey()

# first diagram, indexed by prior = 0
sankey.add(flows=[1, -1], labels=['input', 'output'])

# connect(prior, this)
sankey.add(flows=[1, -1], labels=['input2', 'output2'], prior=0, connect=(1, 0))
sankey.finish()
```

```python
# trunk length of diagram
sankey = Sankey()
sankey.add(flows=[1, -1], labels=['input', 'output'], trunklength=3)
sankey.add(flows=[1, -1], labels=['input2', 'output2'], prior=0, connect=(1, 0))

sankey.finish()
```
